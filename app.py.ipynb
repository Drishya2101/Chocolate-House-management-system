```python
from flask import Flask, request, render_template, redirect, url_for
import sqlite3

app = Flask(__name__)

def get_db_connection():
    conn = sqlite3.connect('chocolate_house.db')
    conn.row_factory = sqlite3.Row
    return conn

@app.route('/')
def index():
    conn = get_db_connection()
    flavors = conn.execute('SELECT * FROM seasonal_flavors').fetchall()
    inventory = conn.execute('SELECT * FROM inventory').fetchall()
    suggestions = conn.execute('''
        SELECT cs.id, sf.name, cs.suggestion, cs.allergy_concerns
        FROM customer_suggestions cs
        JOIN seasonal_flavors sf ON cs.flavor_id = sf.id
    ''').fetchall()
    conn.close()
    return render_template('index.html', flavors=flavors, inventory=inventory, suggestions=suggestions)

@app.route('/add_flavor', methods=['POST'])
def add_flavor():
    name = request.form['name']
    description = request.form['description']
    available_from = request.form['available_from']
    available_to = request.form['available_to']
    
    conn = get_db_connection()
    conn.execute('''
        INSERT INTO seasonal_flavors (name, description, available_from, available_to)
        VALUES (?, ?, ?, ?)
    ''', (name, description, available_from, available_to))
    conn.commit()
    conn.close()
    
    return redirect(url_for('index'))

@app.route('/add_ingredient', methods=['POST'])
def add_ingredient():
    ingredient = request.form['ingredient']
    quantity = request.form['quantity']
    
    conn = get_db_connection()
    conn.execute('''
        INSERT INTO inventory (ingredient, quantity)
        VALUES (?, ?)
    ''', (ingredient, quantity))
    conn.commit()
    conn.close()
    
    return redirect(url_for('index'))

@app.route('/add_suggestion', methods=['POST'])
def add_suggestion():
    flavor_id = request.form['flavor_id']
    suggestion = request.form['suggestion']
    allergy_concerns = request.form['allergy_concerns']
    
    conn = get_db_connection()
    conn.execute('''
        INSERT INTO customer_suggestions (flavor_id, suggestion, allergy_concerns)
        VALUES (?, ?, ?)
    ''', (flavor_id, suggestion, allergy_concerns))
    conn.commit()
    conn.close()
    
    return redirect(url_for('index'))

if __name__ == "__main__":
    app.run(host="0.0.0.0",port=5000)
```
